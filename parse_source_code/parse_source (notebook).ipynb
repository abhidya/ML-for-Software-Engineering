```python
import bblfsh
from collections import defaultdict
import re

#connect to babelfish client
client = bblfsh.BblfshClient("0.0.0.0:9432")
#ast = client.parse("code_samples/%s" % sys.argv[1]).uast
asts_labeled = {}
directory = dirname + '/../StackOverflowScraper/SOfiles/'
for key, value in Name_Code.items():
    filename = directory + key
    ast = client.parse(filename).uast
    asts_labeled[str(ast)] = value
    #uncomment this to look at the AST produced
    print(str(ast))
```

```python
#### MAKING DATAFRAME IN PANDAS ####
import pandas as pd

df = pd.DataFrame(asts_labeled)
df = df.applymap(str)
print(df.columns.values)
```
